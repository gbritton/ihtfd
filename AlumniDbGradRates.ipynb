```python
#!/usr/bin/env python

import numpy as np
import matplotlib.pyplot as plt
from cStringIO import StringIO

# Data taken by hand by searching the Alumni Directory
# Living group is self-reported, as is "Preferred Class Year" (though that is
# at least pre-populated).
csv_data = '''\
2008,75,71,93
2009,46,61,93
2010,44,86,98
2011,44,66,86
2012,103,69,91
2013,59,58,86
2014,91,79,90
2015,94,83,93
2016,108,87,96
2017,26,77,NaN
'''

data = np.loadtxt(StringIO(csv_data), delimiter=',')

fig, ax = plt.subplots(2, sharex=True, gridspec_kw = {'height_ratios':[1, 2]})
ax[0].plot(data[:,0], data[:,1], 'k-')
ax[0].set_title('Senior Haus Graduation Rates (Alumni DB)')
ax[0].set_ylabel('# of students')
ax[0].set_yticks([25,50,75,100,125])
ax[0].grid()

ax[1].plot(data[:,0], data[:,2], 'r-')
ax[1].plot(data[:,0], data[:,3], 'g-')
ax[1].set_ylabel('% 4-year / % 6-year')
ax[1].set_ylim(65,100)
ax[1].set_yticks([50,60,70,80,90,100])
ax[1].set_xticks(data[:,0])
ax[1].set_xlabel('class year')
# 4yr/6yr rates from: https://chancellor.mit.edu/sites/default/files/MITGraduationRates-2008-2015.pdf
ax[1].axhline(59.7, linestyle=':', color='r')
ax[1].axhline(78.1, linestyle=':', color='g')
ax[1].grid()

fig.tight_layout()
plt.show()
```
